```python
from ultralytics import YOLO
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
model=YOLO('yolov8n.pt')
```

```python
cap=cv2.VideoCapture('/content/drive/My Drive/WhatsApp Video 2023-07-22 at 5.03.07 PM(1).mp4')
if not cap.isOpened():
  print('no video')
  exit()
```

```python
while True:
  ret,frame=cap.read()
  if not ret:
    print('unable to read')
    break
  results=model(frame,conf=0.5)
  img=results[0].plot()
  cv2_imshow(img)
cap.release()
cv2.destroyAllWindows()

```
