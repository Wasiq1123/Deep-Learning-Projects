```python
import cv2
from qwen_vl_utils import process_vision_info
from PIL import Image
```

```python
import torch
from transformers import Qwen2_5_VLForConditionalGeneration, AutoProcessor
```

```python
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2.5-VL-3B-Instruct", torch_dtype="auto", device_map="auto"
)
```

```python
cat='/content/drive/My Drive/Clip tokenization dataset/cat.jpg'
img=Image.open(cat)
```

```python
msg = [
    {
        "role": "user",
        "content": [
            {"type": "image", "image": img},
            {"type": "text",  "text": "Describe this image."}
        ],
    }
]
```

```python
processor=AutoProcessor.from_pretrained("Qwen/Qwen2.5-VL-3B-Instruct")
```

```python
text_prompt=processor.apply_chat_template(msg, add_generation_prompt=True, tokenize=False)
```

```python
img_inputs,video_inputs=process_vision_info(msg)
```

```python
inputs=processor(text=[text_prompt],images=img_inputs,return_tensors="pt").to(model.device)
```

```python
with torch.no_grad():
  outputs=model.generate(**inputs,max_new_tokens=100)
```

```python
print(outputs.shape)
```

```python
output=processor.batch_decode(outputs[:inputs.input_ids.shape[-1]:])[0]
print(output)
```
